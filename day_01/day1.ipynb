# Example

## Part 1: example

```python
left = []
right = []

f = open('./example_day1.txt', 'r')
for line in f:
    l, r = line.replace('\n','').split('   ')
    left.append(l)
    right.append(r)

f.close()
left
```

```python
left.sort()
right.sort()

distance = []

for l, r in zip(left,right):
    distance.append(abs(int(l) - int(r)))

sum(distance)
```

## Part 2: example

```python
count = {}
left = []

f = open('./example_day1.txt', 'r')
for line in f:
    l, r = line.replace('\n','').split('   ')

    if r in count:
        count[r] = count[r] + 1
    else:
        count[r] = 1

    left.append(l)

f.close()
count
```

```python
total_left = []

for num in left:
    if num in count:
        total_left.append(int(num)*count[num])
    else:
        total_left.append(0)

sum(total_left)
```
