# Example

```python
zone = []

with open('./example_day08.txt') as f:
    for line in f:
        zone.append([x for x in line.strip()])

zone
```

```python
anti_nodes = np.empty((len(zone), len(zone[0])), dtype=int)
anti_nodes.fill(0)

for i in range(len(zone)):
    for j in range(len(zone[i])):
        if zone[i][j] != '.':
            antena_1 = zone[i][j]

            for ii in range(len(zone)):
                for jj in range(len(zone[ii])):
                    if zone[ii][jj] != '.' and (ii != i or jj != j):
                        antena_2 = zone[ii][jj]
                        if antena_1 == antena_2:
                            node_y = abs(i - ii)
                            node_x = abs(j - jj)

                            if i < ii:
                                node_y = -node_y + i
                            else:
                                node_y = node_y + i
                            if j < jj:
                                node_x = -node_x + j
                            else:
                                node_x = node_x + j

                            if node_y >= 0 and node_y < len(zone) and node_x >= 0 and node_x < len(zone[0]) :
                                anti_nodes[node_y][node_x] = 1

anti_nodes.sum()
```

```python
anti_nodes = np.empty((len(zone), len(zone[0])), dtype=int)
anti_nodes.fill(0)
second_zone = np.copy(zone)

for i in range(len(zone)):
    for j in range(len(zone[i])):
        if zone[i][j] != '.':
            antena_1 = zone[i][j]
            anti_nodes[i][j] = 1

            for ii in range(len(zone)):
                for jj in range(len(zone[ii])):
                    if zone[ii][jj] != '.' and (ii != i or jj != j):
                        antena_2 = zone[ii][jj]
                        if antena_1 == antena_2:
                            y = abs(i - ii)
                            x = abs(j - jj)

                            if i < ii:
                                y = -y 
                            else:
                                y = y
                            if j < jj:
                                x = -x 
                            else:
                                x = x

                            node_y = i + y
                            node_x = j + x

                            while True:
                                if node_y < 0 or node_y >= len(zone) or node_x < 0 or node_x >= len(zone[0]):
                                    break

                                anti_nodes[node_y][node_x] = 1
                                second_zone[node_y][node_x] = '#'

                                node_y = node_y + y
                                node_x = node_x + x

anti_nodes.sum()
```
